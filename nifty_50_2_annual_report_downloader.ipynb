```python
import os
import requests
import pandas as pd

from bs4 import BeautifulSoup

# from selenium import webdriver
import undetected_chromedriver as uc

from time import sleep
%config Completer.use_jedi = False

import warnings
warnings.filterwarnings("ignore")


# !pip install undetected_chromedriver
```

```python
download_loc = "./annual_reports/"

if not os.path.exists(download_loc):
    os.mkdir(download_loc)
```

```python
def annual_reports_downloader(file_link, file_name, file_save_loc, report_year):
    
    location = os.path.join(file_save_loc,report_year)
    if not os.path.exists(location):
        os.mkdir(location)
    location = os.path.join(location,file_name)
    
    if not os.path.exists(location):
        # Downloading the file
        file_content = requests.get(file_link)

        # Saving file
        with open(location,'wb') as output_file:
            output_file.write(file_content.content)
        sleep(5)
```

```python
company_details_df = pd.read_csv("./company_details.csv")
company_details_df
```

```python
print('Downloading started . . .')

# starting chrome driver
driver = uc.Chrome()
```

```python
for i,values in enumerate(company_details_df.values):
    company_name = values[0]
    company_symbol = values[2]
    print("{0:5}. {1:16} :: {2:5}".format(i+1, company_symbol, company_name), end='\t')

    # visiting the webpage
    driver.get('https://www.nseindia.com/companies-listing/corporate-filings-annual-reports')

#     sleep(1)
    # inserting company symbol in search field
    search_field = driver.find_element_by_id('ARCompanyInput')
    search_field.send_keys(company_symbol)
    sleep(1)
    
    file_save_loc = os.path.join(download_loc, company_symbol)
    if not os.path.exists(file_save_loc):
        os.mkdir(file_save_loc)

    search_button = driver.find_element_by_xpath('//*[@id="Annual_Reports_equity"]/div[1]/div[2]/div/button')
    search_button.click()
    sleep(5)

    page_source = BeautifulSoup(driver.page_source)
    reports_info = page_source.findAll('tbody')[0].findAll('tr')
    
    if len(reports_info) > 1:
        print(len(reports_info))
        for count,report in enumerate(reports_info):
            report_year = report.text[-8:-4]
            file_link = report.a['href']
            file_name = file_link.split("/")[-1]
    #         print("\t{0:3}. {1:40} :: {2:1}".format(count+1, file_name, file_link))
            annual_reports_downloader(file_link, file_name, file_save_loc, report_year)
        #     break
    else:
        print("NO REPORT")
    
#     sleep(2)
    
print('Downloading Completed')
```
